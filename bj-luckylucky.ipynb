```python
def deal(n_players, deck):
    players_hands = []
    dealer_hand = []
    
    for i in range(n_players):
        hand = []
        hand.append(deck.pop())
        players_hands.append(hand)
    
    dealer_hand.append(deck.pop())
    
    for i in range(n_players):
        players_hands[i].append(deck.pop())
        
    dealer_hand.append(deck.pop())
    
    return players_hands, dealer_hand, deck
```
